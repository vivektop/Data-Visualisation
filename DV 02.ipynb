```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
telco = pd.read_csv('D:/Datasets\Matplotlib_VisualizationDatasets/Telco-Customer-Churn.csv')
telco.head()
```

#### DONUT chart

```python
pmt_method = telco.groupby(['PaymentMethod']).size()
pmt_method = pd.DataFrame(pmt_method)
pmt_method = pmt_method.reset_index()
pmt_method = pmt_method.rename(columns={0:'Count'})
pmt_method
```

```python
col = ['red','green''yellow','blue']
plt.pie(pmt_method['Count'],labels = pmt_method['PaymentMethod'],labeldistance=0.9,autopct ='%0.1f%%',startangle=90,
        wedgeprops = dict(width = 0.5),radius = 1)

plt.title("Variation by Payment Method",loc = 'center',color = 'maroon')
plt.axis('equal')
plt.show()
```

```python
# DOnut chart with Legends

col = ['red','green','yellow','blue']
pie = plt.pie(pmt_method['Count'],colors=col,autopct ='%0.1f%%',startangle=90,
        wedgeprops = dict(width = 0.5),radius = 1)
plt.title("Variation by Payment Method",loc = 'center',color = 'maroon')
plt.axis('equal')
plt.legend(pie[0],labels = pmt_method['PaymentMethod'],loc ='best',bbox_to_anchor = (1.5,0.7))
```

```python
# Payment by Gender
pmt_by_gdr = telco.groupby(['PaymentMethod','gender']).size()
pmt_by_gdr = pd.DataFrame(pmt_by_gdr)
pmt_by_gdr = pmt_by_gdr.reset_index()
pmt_by_gdr = pmt_by_gdr.rename(columns={0:'Count'})
pmt_by_gdr
```

```python
# 
col = ['coral','crimson','goldenrod','orchid']
col1 = ['steelblue','green','brown','orange']
pie = plt.pie(pmt_by_gdr[pmt_by_gdr['gender'] == 'Male']['Count'],autopct ='%0.1f%%',pctdistance = 0.8,startangle=90,
        wedgeprops = dict(width = 0.5),radius = 1,colors = col)
pie1 = plt.pie(pmt_by_gdr[pmt_by_gdr['gender'] == 'Female']['Count'],autopct ='%0.1f%%',pctdistance = 1.3,startangle=90,
        wedgeprops = dict(width = 0.5),radius = 1.6,colors = col1)
plt.title("Variation by Payment Method",loc = 'center')
plt.legend(pie[0],labels = pmt_by_gdr[pmt_by_gdr['gender']=='Female']['PaymentMethod'],loc ='best',bbox_to_anchor = (1.5,0.7))
plt.legend(pie1[0],labels = pmt_by_gdr[pmt_by_gdr['gender']=='Male']['PaymentMethod'],loc ='right',bbox_to_anchor = (1.9,0.9))
plt.axis('equal')
plt.show()
```

```python
plt.style.use('seaborn-notebook')
```

```python
##Stacked Bar chart

plt.bar(pmt_by_gdr[pmt_by_gdr['gender']=='Male']['PaymentMethod'],
              pmt_by_gdr[pmt_by_gdr['gender']=='Male']['Count'],
             width = 0.35) # x and y axis
plt.bar(pmt_by_gdr[pmt_by_gdr['gender']=='Female']['PaymentMethod'],
                pmt_by_gdr[pmt_by_gdr['gender'] == 'Female']['Count'],
        bottom = pmt_by_gdr[pmt_by_gdr['gender']=='Male']['Count'],
               width = 0.35) # x and y axis
plt.xticks(rotation = 'vertical')
plt.show()

#plt.barh for Horizontal bars and left instead of bottom
```

```python
#plt.style.available

plt.barh(pmt_by_gdr[pmt_by_gdr['gender']=='Male']['PaymentMethod'],
              pmt_by_gdr[pmt_by_gdr['gender']=='Male']['Count']) # x and y axis
plt.barh(pmt_by_gdr[pmt_by_gdr['gender']=='Female']['PaymentMethod'],
                pmt_by_gdr[pmt_by_gdr['gender'] == 'Female']['Count'],
        left = pmt_by_gdr[pmt_by_gdr['gender']=='Male']['Count'])
plt.xticks(rotation = 'vertical')
plt.show()
```
