## Seaborn

```python
import seaborn as sns
tips = sns.load_dataset('tips')
```

```python
tips.shape
```

```python
# Boxplot
sns.barplot(x= 'sex',y = 'total_bill',data = tips)
plt.show()
```

```python
#Barplot
sns.boxplot(x= 'sex',y = 'total_bill',data = tips)
plt.show()
```

```python
# for smoker 
sns.boxplot(x= 'smoker',y = 'total_bill',data = tips)
plt.show()
```

```python
#Hue is for third variable (shade)
sns.boxplot(x='smoker',y = 'total_bill',hue = 'sex',data = tips)
plt.show()
```

```python
sns.boxplot(x= 'sex',y = 'total_bill', hue ='sex', data = tips, palette = 'Pastel2')
plt.show()
```

```python
sns.countplot(x= 'day',data = tips)

#Histogram is for continous data , barplot is discrete
```

```python
sns.violinplot(x='day',y='total_bill', data = tips)
plt.show()

# violin plot shows distribution 
```

```python
sns.violinplot(x='day',y='total_bill',hue = 'sex',data = tips)
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',hue = 'sex',data = tips,split = True) # splits for male and female
plt.show()
```

```python
sns.stripplot(x= 'day',y='total_bill',data = tips) # scatter plot
plt.show()

# If jitter = True
sns.stripplot(x= 'day',y='total_bill',data = tips,jitter = True)
plt.show()

# Using hue(add a 3rd variable) and dodge(is for splitting the smoker and non smoker)
sns.stripplot(x= 'day',y='total_bill',data = tips,jitter = True,hue = 'smoker',dodge = True )
plt.show() # to seperate smoker and non smoker use dodge = True
```

```python
sns.swarmplot(x = 'day',y = 'total_bill',data = tips)
plt.show()
```

### Credit- Default - Assignment in Class

```python
credit = pd.read_excel('D:/Datasets/Matplotlib_VisualizationDatasets/credit-default.xls')
credit.head()
```

```python
# plot the countplot for a) SEX b) EDUCATION c) MARRIAGE d) Default payment next month

# a) SEX

plt.subplot(2,2,1)

sns.countplot(x= 'SEX',data = credit)

# b) Education
plt.subplot(2,2,2)

sns.countplot(x= 'EDUCATION',data = credit)

# c) Marriage
plt.subplot(2,2,3)

sns.countplot(x= 'MARRIAGE',data = credit)

#d) Default payment
plt.subplot(2,2,4)

sns.countplot( x='default payment next month',data = credit)

plt.tight_layout()
plt.show()
```

```python
plt.subplot(2,2,1)
# Parallel Boxplot for 
# a) LIMIT_BAL for each default payment next month level

sns.boxplot(x= 'default payment next month',y = 'LIMIT_BAL',data = credit)

plt.subplot(2,2,2)

# b) AGE for each default payment next month level

sns.boxplot(x= 'default payment next month',y = 'AGE',data = credit)

plt.subplot(2,2,3)

# c) BILL_AMT1 for each default payment next month

sns.boxplot(x= 'default payment next month',y = 'BILL_AMT1',data = credit)

plt.subplot(2,2,4)

# d) BILL_AMT4 for each default payment next month

sns.boxplot(x ='default payment next month',y = 'BILL_AMT4',data = credit )

plt.tight_layout()
plt.show()
```

```python
plt.subplot(1,3,1)

# limit_bal for each education level

sns.boxplot(x ='EDUCATION',y = 'LIMIT_BAL',data = credit)


plt.subplot(1,3,2)

# limit_bal for each marriage level

sns.boxplot(x = 'MARRIAGE',y ='LIMIT_BAL',data = credit)

plt.subplot(1,3,3)

# limit_bal for pay_0 level

sns.boxplot(x = 'PAY_0',y ='LIMIT_BAL',data = credit)

plt.tight_layout()
plt.show()
```

```python
#Parallel boxplot for 
plt.subplot(1,2,1)
sns.boxplot(x = 'SEX',y ='AGE',data = credit,hue = 'default payment next month')
plt.subplot(1,2,2)
sns.boxplot(x = 'SEX',y ='LIMIT_BAL',data = credit,hue = 'default payment next month')
plt.tight_layout()
plt.show()
```

```python
# Plot violinplot for 
plt.subplot(1,2,1)
sns.violinplot(x='default payment next month',y = 'LIMIT_BAL',data = credit,split = True)
plt.subplot(1,2,2)
sns.violinplot(x='default payment next month',y = 'AGE',data = credit,split = True)
plt.tight_layout()
plt.show()
```
